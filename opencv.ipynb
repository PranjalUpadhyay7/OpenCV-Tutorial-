# Image Display in OpenCV

This code demonstrates how to read, display, and close an image window using OpenCV's `imread`, `imshow`, `waitKey`, and `destroyAllWindows` functions.

### 1. Reading the Image

- **`cv2.imread(path, flag)`**: Reads an image from the specified file.

  - **`path`**: The file path to the image. Use a relative path (e.g., `"image.png"`) for files in the same directory, or an absolute path for images located elsewhere.

  - **`flag`**: Determines how the image is read. Options include:
  
    - **`0`** or **`cv2.IMREAD_GRAYSCALE`**: Reads the image in grayscale mode.
    
    - **`1`** or **`cv2.IMREAD_COLOR`**: Reads the image in color (default setting).
    
    - **`-1`** or **`cv2.IMREAD_UNCHANGED`**: Reads the image as is, including any alpha (transparency) channel.

### 2. Displaying the Image in a GUI Window

- **`cv2.imshow(window_title, image)`**: Creates a GUI window to display the specified image.

  - **`window_title`**: The name displayed in the title bar of the window. It should be a string, like `"image"`.

  - **`image`**: The image array to display, typically the output of `cv2.imread()`.

### 3. Keeping the Window Open

- **`cv2.waitKey(delay)`**: Waits for a specified amount of time for the user to press any key.

  - **`delay`**: Time to wait in milliseconds:
  
    - **`0`**: Keeps the window open indefinitely until a key is pressed.
    
    - **Positive number**: Closes the window after the specified number of milliseconds.

### 4. Closing the Window

- **`cv2.destroyAllWindows()`**: Closes all OpenCV windows opened by the `cv2.imshow()` function and frees up memory used by these windows.


````python
import cv2

# to read the image we use 'imread(path, flag)' function. It has two arguments path and flag.
# Path: give relative path for the images in same directory and give full path for the images from other directories.
# Flag: 0, 1 or -1. for grayscale image use 0 or cv2.IMREAD_GRAYSCALE, for color image use 0 or cv2.IMREAD_COLOR,
# for loading images as it is use -1 or cv2.IMREAD_UNCHANGED.
img = cv2.imread("opencv_logo.png", cv2.IMREAD_COLOR)

# Creating GUI window to display an image on screen
# first Parameter is windows title (should be in string format)
# Second Parameter is image array
cv2.imshow("image", img)

# To keep the window open, we use the cv2.waitKey() function.
# It waits for the user to press any key before closing the window.
# The argument is the time to wait, in milliseconds.
# If you pass a positive number (like 5000), it will wait for that many milliseconds.
# If you pass 0, the window will stay open until you press a key to close it.

cv2.waitKey(0)  # Example: 0 means the window stays open until you press a key.
# cv2.waitKey(5000) # you can uncomment this and comment above to see the difference

# It is for removing/deleting created GUI window from screen and memory
cv2.destroyAllWindows()
````

# OpenCV `cv2.imwrite()` Function

The `cv2.imwrite()` function in OpenCV saves an image to a specified file location.

This function is helpful when you want to export images from an OpenCV application, for example, after processing or editing images, directly onto your filesystem.

---

### Parameters

- **filename**:  

  Specifies the complete file path where the image will be saved, including the filename and extension (such as `.png` or `.jpg`). If only the filename is
  
  provided, the image will be saved in the current working directory. It supports
  - JPEG (.jpg, .jpeg)
  
  - PNG (.png)
  
  - TIFF (.tiff, .tif)
  
  - BMP (.bmp)
  
  - PPM (.ppm)
  
  - PGM (.pgm)
  

- **img**:  
  
  Refers to the image array to be saved, typically a result from image processing with OpenCV. This array is usually in the form of a `numpy` array.

- **params** (optional):  
  
  A list of parameters for saving the image in different formats. Examples include:

  - For JPEG images, the `cv2.IMWRITE_JPEG_QUALITY` parameter controls quality on a scale from 0 to 100, with 95 as the default. Higher values yield
  
  better image quality but increase file size.

  - For PNG images, the `cv2.IMWRITE_PNG_COMPRESSION` parameter adjusts compression from 0 to 9, where 3 is the default. Lower values result in less
  
  compression and larger file sizes.

---

### Return Value

The function returns a boolean indicating whether the operation was successful:

- **True**: The image was saved successfully.

- **False**: Indicates an issue, such as an invalid file path or a file system error.


````python
# Save the image with 90% quality
cv2.imwrite("Saved_image_high_quality.jpg", img_rgb, [cv2.IMWRITE_JPEG_QUALITY, 90])

# Save the image with 50% quality (more compressed)
cv2.imwrite("Saved_image_low_quality.jpg", img_rgb, [cv2.IMWRITE_JPEG_QUALITY, 50])


# Save the image with lower compression (higher file size)
cv2.imwrite("Saved_image_low_compression.png", img_rgb, [cv2.IMWRITE_PNG_COMPRESSION, 1])

# Save the image with higher compression (smaller file size)
cv2.imwrite("Saved_image_high_compression.png", img_rgb, [cv2.IMWRITE_PNG_COMPRESSION, 9])
````

# Video Reading using VideoCapture
##Syntax: `cv2.VideoCapture(filename_or_index, apiPreference=None)`
- `filename_or_index`: Path to the video file or index of the camera (e.g., 0 for the default camera).

  - For files, specify the file path. For live capture, provide the camera index.

- `apiPreference` (optional): Preferred capture API.

  - Allows selection of a specific API backend, like `cv2.CAP_FFMPEG`.


````python
import cv2

# Initialize the video capture (0 is usually the default camera)
cap = cv2.VideoCapture(0)

# Check if the camera opened successfully
if not cap.isOpened():
    print("Error: Could not open video source.")
    exit()

# Read and display frames in a loop
while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    # If frame read was successful, display it
    if ret:
        cv2.imshow("Video Feed", frame)
    else:
        print("Error: Failed to capture frame.")
        break

    # Press 'q' to exit the video feed loop
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture object and close display window
cap.release()
cv2.destroyAllWindows()
````

# Writing Video File (Saving)

### Syntax: `cv2.VideoWriter(filename, fourcc, fps, frameSize, isColor=True)`

- `filename`

  - **Type**: `str`

  - **Description**: Path to save the video file, including the filename (e.g., `output.avi`).


- `fourcc`

  - **Type**: `str` (Four-character code)

  - **Description**: Specifies the codec; e.g., `cv2.VideoWriter_fourcc(*'XVID')`.

- `fps`

  - **Type**: `int` or `float`

  - **Description**: Frames per second for the output video.


- `frameSize`

  - **Type**: `tuple` (width, height)

  - **Description**: Dimensions of video frames in pixels (e.g., `(640, 480)`).

- `isColor`

  - **Type**: `bool`

  - **Description**: If `True`, saves video in color; if `False`, saves in grayscale.


| Codec    | Compression Efficiency    | Quality               | Compatibility   | Best Use Case                              |
|----------|---------------------------|-----------------------|-----------------|--------------------------------------------|
| XVID     | Moderate                  | Good                  | Very High       | General-purpose, broad compatibility       |
| H.264    | High                      | Excellent             | Very High       | HD streaming, Blu-ray, online video        |
| H.265    | Very High                 | Optimal               | Moderate        | 4K/8K streaming, storage optimization      |
| MJPEG    | Low                       | Frame-by-frame        | High            | CCTV, video capture                        |
| VP8/VP9  | VP8: High, VP9: Very High | VP8: Good, VP9: Excellent | Moderate  | Web streaming (YouTube)                    |
| AV1      | Very High                 | Excellent             | Growing         | Future-proof streaming, HD+ web content    |
| ProRes   | Moderate                  | Optimal               | Apple-centric   | Professional video editing, broadcast      |


````python
import cv2

# Specify the path to save the video file
output_path = 'output_video_120fps.avi'

# Set up video capture (replace 0 with your video source or file path)
cap = cv2.VideoCapture(0)

# Define video properties
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
frame_rate = 120  # Set frame rate to 120 FPS
# # Specify codec using fourcc for XVID
# fourcc = cv2.VideoWriter_fourcc(*'XVID')

# # Specify codec using fourcc for H.264
# fourcc = cv2.VideoWriter_fourcc(*'H264')

# # Specify codec using fourcc for MJPEG
# fourcc = cv2.VideoWriter_fourcc(*'MJPG')

# Define the codec and create a VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter(output_path, fourcc, frame_rate, (frame_width, frame_height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Write the frame to the output video
    out.write(frame)

    # Display the frame (optional)
    cv2.imshow('Frame', frame)

    # Press 'q' to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
out.release()
cv2.destroyAllWindows()
````

# Arithmetic Operations on Images Using OpenCV

- Arithmetic operations such as addition, subtraction, and bitwise operations (AND, OR, NOT, XOR) can be performed on input images using OpenCV.

These operations are beneficial for enhancing the properties of the images.

- Image arithmetic is crucial for analyzing various characteristics of the input images. The results of these operations can serve as enhanced inputs for

further processing, including tasks like clarification, thresholding, and dilation.

## Addition
The `cv2.addWeighted()` function is designed to compute the weighted sum of two images, making it particularly useful for tasks like image blending and

adjusting brightness and contrast.

### Parameters

- **img1**:  The first input image array (single-channel, 8-bit or floating-point).

- **wt1**:  The weight of the first input image elements to be applied to the final output image.

- **img2**:  The second input image array (single-channel, 8-bit or floating-point).

- **wt2**:  The weight of the second input image elements to be applied to the final output image.

- **gammaValue**:  A measurement of light that is added to the weighted sum. This parameter can be used to adjust the brightness of the final result.

### Formula

The function computes the final pixel values using the following formula:

## output = img1 X wt1 + img2 X wt2 + gammaValue



````python

# Read the images
image1_add = cv2.imread('Add1.jpg')
image2_add = cv2.imread('Add2.jpg')

# Apply cv2.addWeighted to get the weighted sum of the images
weightedSum = cv2.addWeighted(image1_add, 0.5, image2_add, 0.4, 0)

# # Convert the image from BGR to RGB format for displaying with matplotlib
weightedSum = cv2.cvtColor(weightedSum, cv2.COLOR_BGR2RGB)
concatenated_images = np.hstack((image1_add, image2_add, weightedSum))

# Display the concatenated images in a single window
cv2.imshow('Image 1 | Image 2 | Added', concatenated_images)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

## Subtraction

The `cv2.subtract()` function performs per-element subtraction of two images, where each pixel in the output image is computed by subtracting the

corresponding pixel values of two input images. This operation is useful in applications like background subtraction, change detection, or highlighting

differences between two images.

### Parameters

- **image1_sub**: The first input image array (single-channel, 8-bit, or floating-point).

- **image2_sub**: The second input image array (single-channel, 8-bit, or floating-point).


````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Read the images
image1_sub = cv2.imread('sub1.jpg')
image2_sub = cv2.imread('sub2.jpg')

# Convert images to RGB (from BGR) for displaying with matplotlib
image1_rgb = cv2.cvtColor(image1_sub, cv2.COLOR_BGR2RGB)
image2_rgb = cv2.cvtColor(image2_sub, cv2.COLOR_BGR2RGB)

# Subtract the images
subtracted_img = cv2.subtract(image1_sub, image2_sub)
subtracted_img_rgb = cv2.cvtColor(subtracted_img, cv2.COLOR_BGR2RGB)

# Display images in subplots
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

ax[0].imshow(image1_rgb)
ax[0].set_title('Image 1')
ax[0].axis('off')

ax[1].imshow(image2_rgb)
ax[1].set_title('Image 2')
ax[1].axis('off')

ax[2].imshow(subtracted_img_rgb)
ax[2].set_title('Subtracted')
ax[2].axis('off')

# Show the plot
plt.show(block=False)

````

## Bitwise Operations

Bitwise operations allow pixel-wise manipulation of images, where each pixel in the output image is calculated based on the binary representation of pixel

values in the input images. OpenCV provides functions for various bitwise operations, including AND, OR, XOR, and NOT. These operations can be useful for

tasks like masking, segmentation, and image enhancement.

### Parameters

- **img1**: The first input image array (single or multi-channel, typically 8-bit or floating-point).

- **img2**: The second input image array, which should be the same size as `img1`.

### Bitwise AND

The `cv2.bitwise_and()` function computes the per-element logical AND between two images, where each pixel in the output is the result of `img1 AND img2`.

This operation is useful for highlighting overlapping regions between two images.

### Bitwise OR

The `cv2.bitwise_or()` function performs the per-element logical OR operation between two images, where each output pixel is calculated as `img1 OR img2`.

This operation is often used for merging regions of two images.

### Bitwise XOR

The `cv2.bitwise_xor()` function computes the per-element logical XOR operation between two images, where each output pixel is the result of `img1 XOR img2`.

 It is useful for highlighting non-overlapping regions between two images.

### Bitwise NOT

The `cv2.bitwise_not()` function inverts the pixel values of an image, transforming light areas to dark and vice versa. This operation is commonly used

for creating masks and enhancing contrast.

### Mask Creation

In addition to these operations, a binary mask is created to selectively apply bitwise operations to specific regions of the images.

- **mask**: A single-channel array with the same width and height as `img1`. Non-zero values in the mask retain the operation's effect in the

corresponding output region, while zero values discard the operation in those regions.

### Bitwise OR with Mask

The `cv2.bitwise_or()` function can also be used with a mask, which limits the OR operation to specific regions defined by the mask. This masked operation

is useful for selectively modifying parts of an image based on a region of interest.

### Display

The code converts each output image from BGR to RGB for compatibility with Matplotlib and displays all images in subplots.


````python
# Load images
img1 = cv2.imread('bit1.png')
img2 = cv2.imread('bit2.png')

# Ensure images are loaded
if img1 is None or img2 is None:
    raise ValueError("One of the images couldn't be loaded. Check the file paths.")

# Resize img2 to match img1 if necessary (assuming they should be the same size)
if img1.shape != img2.shape:
    img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

# Convert images from BGR to RGB for matplotlib display
img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

# Perform bitwise operations
bit_and = cv2.bitwise_and(img1, img2)
bit_or = cv2.bitwise_or(img1, img2)
bit_xor = cv2.bitwise_xor(img1, img2)
bit_not1 = cv2.bitwise_not(img1)
bit_not2 = cv2.bitwise_not(img2)

# Create a mask and ensure it's the same size as img1
mask = np.zeros(img1.shape[:2], dtype=np.uint8)  # Single-channel mask
cv2.circle(mask, (mask.shape[1] // 2, mask.shape[0] // 2), 100, 255, -1)
plt.imshow(mask, cmap= 'gray')
plt.title('MASK')
# Apply bitwise AND with mask
masked_or = cv2.bitwise_or(img1, img2, mask=mask)

# Convert results to RGB for matplotlib display
bit_and_rgb = cv2.cvtColor(bit_and, cv2.COLOR_BGR2RGB)
bit_or_rgb = cv2.cvtColor(bit_or, cv2.COLOR_BGR2RGB)
bit_xor_rgb = cv2.cvtColor(bit_xor, cv2.COLOR_BGR2RGB)
bit_not1_rgb = cv2.cvtColor(bit_not1, cv2.COLOR_BGR2RGB)
bit_not2_rgb = cv2.cvtColor(bit_not2, cv2.COLOR_BGR2RGB)
masked_and_rgb = cv2.cvtColor(masked_or, cv2.COLOR_BGR2RGB)

# Plot all images in subplots with a custom background
fig, ax = plt.subplots(3, 3, figsize=(15, 15))
fig.suptitle("Bitwise Operations", fontsize=16)
fig.patch.set_facecolor('lightgray')  # Set figure background color


# Display original images
ax[0, 0].imshow(img1_rgb)
ax[0, 0].set_title('Image 1')
ax[0, 0].axis('off')

ax[0, 1].imshow(img2_rgb)
ax[0, 1].set_title('Image 2')
ax[0, 1].axis('off')

ax[0, 2].imshow(mask, cmap='gray')
ax[0, 2].set_title('Mask')
ax[0, 2].axis('off')

# Display bitwise operation results
ax[1, 0].imshow(bit_and_rgb)
ax[1, 0].set_title('Bitwise AND')
ax[1, 0].axis('off')

ax[1, 1].imshow(bit_or_rgb)
ax[1, 1].set_title('Bitwise OR')
ax[1, 1].axis('off')

ax[1, 2].imshow(bit_xor_rgb)
ax[1, 2].set_title('Bitwise XOR')
ax[1, 2].axis('off')

# Display NOT and masked results
ax[2, 0].imshow(bit_not1_rgb)
ax[2, 0].set_title('Bitwise NOT Image 1')
ax[2, 0].axis('off')

ax[2, 1].imshow(bit_not2_rgb)
ax[2, 1].set_title('Bitwise NOT Image 2')
ax[2, 1].axis('off')

ax[2, 2].imshow(masked_and_rgb)
ax[2, 2].set_title('Bitwise OR with Mask')
ax[2, 2].axis('off')

# Show the plot
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
````

# Resizing:

`cv2.resize(src, dsize, fx=0, fy=0, interpolation=cv2.INTER_LINEAR)`


### Parameters:


- **src**:

  - Type: `ndarray`

  - Description: The source image to be resized.


- **dsize**:

  - Type: `tuple`

  - Description: A tuple specifying the desired size of the output image in the format (width, height). If set to (0, 0), resizing will be done using fx and fy instead.


- **fx**:

  - Type: `float`

  - Description: Scale factor along the x (width) axis. This parameter is optional and used only if dsize is set to (0, 0).


- **fy**:

  - Type: `float`

  - Description: Scale factor along the y (height) axis. This parameter is optional and used only if dsize is set to (0, 0).


- **interpolation**:

  - Type: `int`

  - Description: (Optional) The interpolation method used to resize the image. Common options include:

    - `cv2.INTER_NEAREST`: Nearest-neighbor interpolation, which is fast but may produce blocky images.

    - `cv2.INTER_LINEAR`: Bilinear interpolation (default), which is a good balance between quality and speed.

    - `cv2.INTER_CUBIC`: Bicubic interpolation, which provides better quality for larger enlargements.

    - `cv2.INTER_LANCZOS4`: Lanczos interpolation, which is ideal for high-quality downsampling.


| Method        | Parameters           | Purpose                                   | Example Use Case                          |
|---------------|----------------------|-------------------------------------------|-------------------------------------------|
| Scale Factors | `fx`, `fy`           | Resize by a ratio relative to the original. | Scaling down by 50% or up by 200%.       |
| Exact Size    | `dsize` (width, height) | Resize to specific pixel dimensions.       | Fitting an image to an exact resolution. |


bicubic=https://www.paulinternet.nl/?page=bicubic

lanczos=https://www.youtube.com/watch?v=Xj129kA3Ci0

````python
import cv2
import matplotlib.pyplot as plt

# Load the image
original = cv2.imread('opencv_logo.png')
print(original.shape)
# Resize images to the specified sizes
half = cv2.resize(original,(0,0), fx=0.5, fy=0.5)  # Halving the resized original
bigger = cv2.resize(original, (500, 400))  # Enlarged size
stretch_near = cv2.resize(original, (300, 300), interpolation=cv2.INTER_LINEAR)  # Linear interpolation

# Convert images to RGB for Matplotlib
original_rgb = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)
half_rgb = cv2.cvtColor(half, cv2.COLOR_BGR2RGB)
bigger_rgb = cv2.cvtColor(bigger, cv2.COLOR_BGR2RGB)
stretch_near_rgb = cv2.cvtColor(stretch_near, cv2.COLOR_BGR2RGB)

# Titles and images list
Titles = ["Original (200x250)", "Half", "Bigger", "Interpolation Linear"]
images = [original_rgb, half_rgb, bigger_rgb, stretch_near_rgb]

# Create figure with a gray background
fig = plt.figure(figsize=(8, 8))
fig.patch.set_facecolor('lightgray')  # Set background color to light gray

# Plot images in subplots
for i in range(len(images)):
    ax = fig.add_subplot(2, 2, i + 1)
    ax.set_title(Titles[i])
    ax.imshow(images[i])
# Adjust layout and show the plot
plt.tight_layout()
plt.show()
````

# Applying Geometric Transformations using `cv2.warpAffine()`

`cv2.warpAffine()` is a powerful function in OpenCV that can perform various affine transformations, such as translation, rotation, scaling, and shearing. 

The type of transformation applied to the image depends on the values in the transformation matrix `M`.


## Syntax


`cv2.warpAffine(src, M, dsize, dst=None, flags=cv2.INTER_LINEAR, borderMode=cv2.BORDER_CONSTANT, borderValue=0)`

## Parameters

- **`src`**

  - **Type**: `numpy.ndarray`

  - **Description**: The input image to be transformed.

- **`M`**

  - **Type**: `numpy.ndarray` (2x3 matrix)

  - **Description**: The 2x3 affine transformation matrix that defines the transformation. Different matrix values allow different transformations:

    - **Translation**: Moves the image by shifting pixels along the x and y axes.
      ```

      M = [
        [1, 0, tx],
        [0, 1, ty]
      ]
      ```
    - **Rotation**: Rotates the image around a specified center.
      ```
      M = [
        [cos(θ), -sin(θ), tx],
        [sin(θ), cos(θ), ty]
      ]
      ```
      Use `cv2.getRotationMatrix2D(center, angle, scale)` to get this matrix.
  
    - **Scaling**: Resizes the image along the x and y axes.
      ```
      M = [
        [sx, 0, tx],
        [0, sy, ty]
      ]
      ```
    - **Shearing (Skewing)**: Slants the image, creating a skew effect.
      ```
      M = [
        [1, shx, tx],
        [shy, 1, ty]
      ]
      ```

- **`dsize`**

  - **Type**: `tuple` (width, height)

  - **Description**: Size of the output image in pixels (width, height).

- **`dst`** (optional)

  - **Type**: `numpy.ndarray`

  - **Description**: Destination array where the result is stored. If not provided, a new array is created.

- **`flags`** (optional)

  - **Type**: `int`

  - **Description**: Interpolation method. Options include:

### Image Blurring

**Image Blurring** reduces the clarity and detail of an image, typically using low-pass filter kernels. This technique offers several advantages:

**Noise Reduction**: Blurring helps in noise removal, as noise often contains high-frequency components. Applying a low-pass filter kernel limits the high-frequency noise.

**Smoothing**: Blurring smooths the image, making it visually softer.

**Edge Suppression**: Low-intensity edges are diminished, enhancing the prominence of more significant edges.

**Detail Concealment**: Blurring can obscure details when needed, such as hiding a person’s identity in certain images.

### Types of Blurring

#### Gaussian Blurring

Gaussian blur applies a Gaussian function across the image to produce a smooth, blurred effect. This method is popular in graphics software to reduce

image noise and detail. It's also used as a preprocessing step in machine learning or deep learning.


#### Median Blurring

The **Median Filter** is a non-linear filter that removes noise by replacing each pixel with the median value of its neighboring pixels. This technique is

highly effective for removing "salt and pepper" noise and is known for preserving edges.

#### Bilateral Blurring

**Bilateral Filter** is a non-linear, edge-preserving, noise-reducing smoothing filter. It replaces the intensity of each pixel with a weighted average of

nearby pixels, with weights based on a Gaussian distribution. This approach maintains sharp edges while discarding weaker ones, making it ideal for

edge-sensitive tasks.


### 1. Gaussian Blur

**Function Syntax:**

``cv2.GaussianBlur(src, ksize, sigmaX)``

**Arguments:**

- **`src`:**  

  - Type: `ndarray`  

  - Description: The input source image to be blurred.

- **`ksize`:**  

  - Type: `tuple (width, height)`  

  - Description: The size of the Gaussian kernel. Both width and height should be odd integers (e.g., (7, 7)). Larger values create a stronger blur.

- **`sigmaX`:**  

  - Type: `float`  

  - Description: The standard deviation in the X direction for the Gaussian kernel. Setting `sigmaX` to 0 allows OpenCV to automatically calculate it

  based on `ksize`.

---

### 2. Median Blur

**Function Syntax:**

``cv2.medianBlur(src, ksize)``

**Arguments:**

- **`src`:**  

  - Type: `ndarray`  

  - Description: The input source image to apply median blurring.

- **`ksize`:**  

  - Type: `int`  

  - Description: The aperture linear size of the kernel, representing the neighborhood of pixels used. It should be a positive odd integer (e.g., 5).
  
  Larger values increase the extent of the blurring effect.

---

### 3. Bilateral Blur

**Function Syntax:**

``cv2.bilateralFilter(src, d, sigmaColor, sigmaSpace)``

**Arguments:**

- **`src`:**  

  - Type: `ndarray`  

  - Description: The input source image to be blurred.

- **`d`:**  

  - Type: `int`  

  - Description: The diameter of each pixel neighborhood used during filtering. Larger values increase the number of neighboring pixels considered.

- **`sigmaColor`:**  

  - Type: `float`  

  - Description: Filter sigma in the color space. Higher values allow a larger range of colors to mix, creating a smoother effect.

- **`sigmaSpace`:**  

  - Type: `float`  

  - Description: Filter sigma in the coordinate space. Higher values let pixels farther from the center of the neighborhood influence the blur, helping
  
  preserve edges.


````python
# Importing libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('opencv_logo.png')


# Apply different blurring techniques
# Gaussian Blur
Gaussian = cv2.GaussianBlur(image, (7, 7), 0)
# Median Blur
median = cv2.medianBlur(image, 7)
# Bilateral Blur
bilateral = cv2.bilateralFilter(image, 15, 75, 75)

# Convert images to RGB for Matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
Gaussian_rgb = cv2.cvtColor(Gaussian, cv2.COLOR_BGR2RGB)
median_rgb = cv2.cvtColor(median, cv2.COLOR_BGR2RGB)
bilateral_rgb = cv2.cvtColor(bilateral, cv2.COLOR_BGR2RGB)

# Titles and images list
Titles = ["Original", "Gaussian Blur", "Median Blur", "Bilateral Blur"]
images = [image_rgb, Gaussian_rgb, median_rgb, bilateral_rgb]

# Create figure with a gray background
fig = plt.figure(figsize=(8, 8))
fig.patch.set_facecolor('lightgray')  # Set background color to light gray

# Plot images in subplots
for i in range(len(images)):
    ax = fig.add_subplot(2, 2, i + 1)
    ax.set_title(Titles[i])
    ax.imshow(images[i])
    ax.axis('off')  # Hide axes for a cleaner display

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
````

# Morphological Operations for RGB Images

### `cv2.erode()`

`cv2.erode()` is used to erode (shrink) the white regions in a binary or RGB image. In the case of an RGB image, the operation is applied independently on 

each color channel (Red, Green, and Blue).



**Syntax**:  


`cv2.erode(src, kernel, iterations=1)`



**Arguments**:

- **`src`**  

  - **Type**: `numpy.ndarray`  

  - **Explanation**: The input image on which erosion is applied. For RGB images, the operation will be performed separately on each channel (Red, Green, 

  Blue).
  
- 
**`kernel`**  

  - **Type**: `numpy.ndarray`  

  - **Explanation**: A structuring element used to process the image. It determines how the erosion operation is applied. The kernel can be any shape 

  (typically square, elliptical, or cross-shaped), and its size affects the erosion strength.



- **`iterations`** (optional)  

  - **Type**: `int`  

  - **Explanation**: The number of times the erosion operation is applied. A larger value increases the shrinking effect. Default value is `1`.


---



### `cv2.dilate()`

`cv2.dilate()` is used to dilate (expand) the white regions in a binary or RGB image. As with erosion, the operation will be applied to each color channel 

separately in an RGB image.



**Syntax**:  

`cv2.dilate(src, kernel, iterations=1)`



**Arguments**:

- **`src`**  

  - **Type**: `numpy.ndarray`  

  - **Explanation**: The input image on which dilation is applied. In the case of RGB images, dilation is applied to each channel individually (Red, 

  Green, Blue).



- **`kernel`**  

  - **Type**: `numpy.ndarray`  

  - **Explanation**: The structuring element used to process the image. The kernel defines the shape and size of the area used to expand the white regions 

  in the image.

- 
**`iterations`** (optional)  

  - **Type**: `int`  

  - **Explanation**: The number of times the dilation operation is applied. Default is `1`, and increasing iterations will result in greater expansion. 

  Default value is `1`.



---


### `cv2.morphologyEx()`

`cv2.morphologyEx()` is used for applying advanced morphological operations such as opening, closing, gradient, top hat, and black hat. For RGB images, 

this operation is performed separately on each color channel.



**Syntax**:  

`cv2.morphologyEx(src, op, kernel, iterations=1)`



**Arguments**:

- **`src`**  

  - **Type**: `numpy.ndarray`  

  - **Explanation**: The input image on which the morphological operation is applied. Typically, a binary or RGB image. In the case of RGB, each channel 

  (Red, Green, and Blue) will be processed separately.



- **`op`**  

  - **Type**: `int`  

  - **Explanation**: The specific morphological operation to be performed. Options include:

    - `cv2.MORPH_OPEN`: Opening (erosion followed by dilation)

    - `cv2.MORPH_CLOSE`: Closing (dilation followed by erosion)

    - `cv2.MORPH_GRADIENT`: Difference between dilation and erosion

    - `cv2.MORPH_TOPHAT`: Difference between the original image and the opening result

    - `cv2.MORPH_BLACKHAT`: Difference between the closing result and the original image



- **`kernel`**  

  - **Type**: `numpy.ndarray`  

  - **Explanation**: The structuring element (kernel) used for the operation. It defines the size and shape of the neighborhood used for the operation.



- **`iterations`** (optional)  

  - **Type**: `int`  

  - **Explanation**: The number of times the morphological operation is applied. Default value is `1`. Increasing this value will amplify the effect of 
  the operation.





````python

import cv2

import numpy as np
import matplotlib.pyplot as plt

# Load the input image (make sure to replace the path with your actual image)
img = cv2.imread('opencv_logo.png')
open=cv2.imread('open_morph.png')
close=cv2.imread('close_morph.png')
# Convert the image to RGB (OpenCV loads images in BGR by default)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Define kernel for morphological operations
kernel = np.ones((5, 5), np.uint8)

# Erosion
eroded = cv2.erode(img_rgb, kernel, iterations=1)

# Dilation
dilated = cv2.dilate(img_rgb, kernel, iterations=1)

# Morphological Opening (erosion followed by dilation)
opened = cv2.morphologyEx(open, cv2.MORPH_OPEN, kernel)

# Morphological Closing (dilation followed by erosion)
closed = cv2.morphologyEx(close, cv2.MORPH_CLOSE, kernel)

# Titles and images list
Titles = ["Original", "Eroded", "Dilated"]
title=[ 'Originial_open','Opened', "Original Closed", "Closed"]
images = [img_rgb, eroded, dilated]
image=[open,opened,close,closed]

# Create figure with a gray background
fig = plt.figure(figsize=(10, 10))
fig.patch.set_facecolor('lightgray')  # Set background color to light gray

# Plot images in subplots
for i in range(len(images)):
    ax = fig.add_subplot(1, 3, i + 1)
    ax.set_title(Titles[i])
    ax.imshow(images[i])

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
fig = plt.figure(figsize=(10, 10))
fig.patch.set_facecolor('lightgray')
# Plot images in subplots
for i in range(len(image)):
    ax = fig.add_subplot(2, 2, i+ 1)
    ax.set_title(title[i])
    ax.imshow(image[i])

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
````

### 4. **`cv2.Sobel()`** - Apply the Sobel Edge Detector

     - `src`: Image source.

### 6. **`cv2.Canny()`** - Apply the Canny Edge Detector

   - **Syntax**: `dst = cv2.Canny(image, threshold1, threshold2, apertureSize, L2gradient)`

   - **Parameters**:

     - `image`: Image source (single-channel, grayscale).

     - `threshold1`: Lower threshold for hysteresis.

     - `threshold2`: Upper threshold for hysteresis.

     - `apertureSize`: Aperture size for the Sobel operator (optional).


     - `L2gradient`: Boolean indicating whether to use L2-norm or L1-norm for gradient calculation.


````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image in grayscale
image = cv2.imread('opencv_logo.png', cv2.IMREAD_GRAYSCALE)

# Apply different filters and store results with their titles
filters = [
    ("Original Image", image),
    ("Fixed Threshold", cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)[1]),
    ("Adaptive Threshold", cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)),
    ("Custom Filter", cv2.filter2D(image, -1, np.ones((5, 5), np.float32) / 25)),
    ("Sobel Edge Detection", cv2.magnitude(cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=5), cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5))),
    ("Laplacian Edge Detection", cv2.Laplacian(image, cv2.CV_64F)),
    ("Canny Edge Detection", cv2.Canny(image, 100, 200))
]

# Plot results
fig=plt.figure(figsize=(12, 8))
fig.patch.set_facecolor('lightgray')
for i, (title, result) in enumerate(filters):
    plt.subplot(2, 4, i + 1)
    plt.imshow(result, cmap='gray')
    plt.title(title)
    plt.axis('off')

plt.tight_layout()
plt.show()
````

````python
import cv2
import numpy as np

# Initialize the camera
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Could not open camera.")
    exit()

# Define the filter name (Change this to switch between filters)
filter_name = "canny"  # Choose filter: "adaptive_threshold", "fixed_threshold", "sobel", "laplacian", "canny", "custom_filter"

# Define filter functions
def apply_fixed_threshold(gray_frame):
    _, thresh_frame = cv2.threshold(gray_frame, 127, 255, cv2.THRESH_BINARY)
    return thresh_frame

def apply_adaptive_threshold(gray_frame):
    return cv2.adaptiveThreshold(gray_frame, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

def apply_sobel(gray_frame):
    sobelx = cv2.Sobel(gray_frame, cv2.CV_64F, 1, 0, ksize=5)
    sobely = cv2.Sobel(gray_frame, cv2.CV_64F, 0, 1, ksize=5)
    sobel = cv2.magnitude(sobelx, sobely)
    return cv2.convertScaleAbs(sobel)

def apply_laplacian(gray_frame):
    laplacian = cv2.Laplacian(gray_frame, cv2.CV_64F)
    return cv2.convertScaleAbs(laplacian)

def apply_canny(gray_frame):
    return cv2.Canny(gray_frame, 50, 100)

def apply_custom_filter(gray_frame):
    kernel = np.ones((5, 5), np.float32) / 25
    return cv2.filter2D(gray_frame, -1, kernel)

# Filter mapping
filters = {
    "fixed_threshold": apply_fixed_threshold,
    "adaptive_threshold": apply_adaptive_threshold,
    "sobel": apply_sobel,
    "laplacian": apply_laplacian,
    "canny": apply_canny,
    "custom_filter": apply_custom_filter
}

# Main loop to apply selected filter on live feed
while True:
    ret, frame = cap.read()

    if not ret:
        print("Error: Failed to grab frame.")
        break

    # Convert to grayscale
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Apply the selected filter
    filtered_frame = filters.get(filter_name, apply_adaptive_threshold)(gray_frame)  # Default to adaptive_threshold if filter not found

    # Show the filtered frame
    cv2.imshow("Filtered Frame", filtered_frame)

    # Exit on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
````

# Video Analysis

### 1. `cv2.calcOpticalFlowPyrLK()`

**Syntax**:

```python
cv2.calcOpticalFlowPyrLK(prevImg, nextImg, prevPts, nextPts, winSize, maxLevel, criteria)
```



**Parameters**:

- `prevImg`: First (previous) image in grayscale.

- `nextImg`: Second (next) image in grayscale.

- `prevPts`: Array of points (features) in `prevImg` to track.

- `nextPts`: Output array of tracked points in `nextImg`.

- `winSize`: Size of the search window at each pyramid level.

- `maxLevel`: Maximum pyramid level.

- `criteria`: Termination criteria (e.g., `(cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 10, 0.03)`).




---



### 2. `cv2.calcOpticalFlowFarneback()`



**Syntax**:

```python

cv2.calcOpticalFlowFarneback(prevImg, nextImg, flow, pyr_scale, levels, winsize, iterations, polyN, polySigma, flags)

```



**Parameters**:

- `prevImg`: Previous frame (grayscale).

- `nextImg`: Current frame (grayscale).

- `flow`: Initial flow; set to `None` to compute from scratch.

- `pyr_scale`: Pyramid scale, specifies the image scale (<1) to build pyramids.

- `levels`: Number of pyramid layers.

- `winsize`: Window size for the averaging area.

- `iterations`: Number of iterations for each pyramid level.

- `polyN`: Size of the pixel neighborhood.

- `polySigma`: Standard deviation for Gaussian.

- `flags`: Operation flags (e.g., `cv2.OPTFLOW_FARNEBACK_GAUSSIAN` for a Gaussian window).



---



### 3. `cv2.createBackgroundSubtractorMOG2()`



**Syntax**:

```python

cv2.createBackgroundSubtractorMOG2(history, varThreshold, detectShadows)

```



**Parameters**:

- `history`: Number of frames for background model history.

- `varThreshold`: Threshold for classifying a pixel as foreground.

- `detectShadows`: Boolean; `True` to detect shadows, `False` otherwise.



---





- **Lucas-Kanade**: Sparse optical flow for tracking features.

- **Farneback**: Dense optical flow for calculating per-pixel flow.

- **Background Subtraction**: Distinguishes foreground from background.

````python
import cv2
import numpy as np

# Initialize the camera
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("Error: Could not open camera.")
    exit()

# Choose the method to apply ("lucas_kanade", "farneback", "background_subtraction")
method = "background_subtraction"  # Change to "farneback" or "background_subtraction" for other methods

# Parameters for Lucas-Kanade Optical Flow
lk_params = dict(winSize=(15, 15),
                 maxLevel=2,
                 criteria=(cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 10, 0.03))

# Parameters for Farneback Optical Flow
farneback_params = dict(pyr_scale=0.5, levels=3, winsize=15, iterations=3, poly_n=5, poly_sigma=1.2, flags=0)

# Initialize background subtractor
back_sub = cv2.createBackgroundSubtractorMOG2()

# Initialize variables for Lucas-Kanade
ret, prev_frame = cap.read()
prev_gray = cv2.cvtColor(prev_frame, cv2.COLOR_BGR2GRAY)
p0 = cv2.goodFeaturesToTrack(prev_gray, mask=None, maxCorners=100, qualityLevel=0.3, minDistance=7, blockSize=7)

# Main loop
while True:
    ret, frame = cap.read()
    if not ret:
        print("Error: Failed to grab frame.")
        break

    # Convert to grayscale
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    if method == "lucas_kanade":
        # Calculate Optical Flow using Lucas-Kanade
        if p0 is not None:
            p1, st, err = cv2.calcOpticalFlowPyrLK(prev_gray, gray_frame, p0, None, **lk_params)
            if p1 is not None:
                good_new = p1[st == 1]
                good_old = p0[st == 1]
                for i, (new, old) in enumerate(zip(good_new, good_old)):
                    a, b = new.ravel()
                    c, d = old.ravel()
                    frame = cv2.line(frame, (int(a), int(b)), (int(c), int(d)), (0, 255, 0), 2)
                    frame = cv2.circle(frame, (int(a), int(b)), 5, (0, 0, 255), -1)
            prev_gray = gray_frame.copy()
            p0 = good_new.reshape(-1, 1, 2)

    elif method == "farneback":
        # Calculate Dense Optical Flow using Farneback
        flow = cv2.calcOpticalFlowFarneback(prev_gray, gray_frame, None, **farneback_params)
        mag, ang = cv2.cartToPolar(flow[..., 0], flow[..., 1])
        hsv = np.zeros_like(frame)
        hsv[..., 1] = 255
        hsv[..., 0] = ang * 180 / np.pi / 2
        hsv[..., 2] = cv2.normalize(mag, None, 0, 255, cv2.NORM_MINMAX)
        frame = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)
        prev_gray = gray_frame

    elif method == "background_subtraction":
        # Background subtraction
        fg_mask = back_sub.apply(frame)
        frame = cv2.bitwise_and(frame, frame, mask=fg_mask)

    # Show the processed frame
    cv2.imshow("Processed Frame", frame)

    # Exit on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
````
